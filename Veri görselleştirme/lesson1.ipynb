```python
#Kategorik değişken --> sütun grafik
#Sayısal değişken --> hist, boxplot


import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option("display.max_columns",None)
df = sns.load_dataset("titanic")
df.head()
```

```python

#Kategorik değişken görselleştirme 
df["sex"].value_counts().plot(kind="bar")
plt.show()
```

```python
#sayısal değişken görselleştirme
plt.hist(df["age"])
plt.show()
```

```python
plt.boxplot(df["fare"])
plt.show()
#aykırı değerleri boxplot ile işaretleryebiliriz
```

```python
#plot
x = np.array([1,8])
y = np.array([0,150])

plt.plot(x,y,"o")
plt.show()
```

```python
#marker

y = np.array([13,28,11,100])
plt.plot(y, marker="*")
plt.show()
```

```python
#line
y = np.array([13,28,11,100])
plt.plot(y,linestyle="dashed",color="r")
plt.show()
```

```python
x = np.array([80,85,90,95,100,105])
y = np.array([240,250,260,270,280,290])
plt.plot(x,y)
plt.title("This main title")
plt.xlabel("This x name")
plt.ylabel("This y name")
plt.grid()
plt.show()
```

```python
#subplots
x = np.array([80,85,90,95,100,105])
y = np.array([240,250,260,270,280,290])
plt.subplot(1,2,1)#(1 columns 2 rows)
plt.title("1")
plt.plot(x,y)
plt.show()
```

```python
#iloc and loc 

#iloc (rows,columns)
df.iloc[0:3] #it will choose 0,1,2
df.iloc[0:3,0:3] #it will show 0 to 3 rows and 0 to 3 columns

df.loc[0:3] #it will choose 0,1,2,3 
df.loc[0:3, "age"]
```

```python
col_names=["age","embarked","alive"]
df.loc[0:3, col_names]
```

```python
#we want to choose that older than age of 50

df[df["age"]>50].head()
df[df["age"]>50]["age"].count()
```
